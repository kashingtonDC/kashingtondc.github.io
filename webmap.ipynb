```python
import os
import leafmap.foliumap as leafmap
```

```python
vis = '/composites/KanabRiver_vis.tif'
tir = '/composites/KanabRiver_tir.tif'
```

```python
m = leafmap.Map(draw_control=True,google_map="HYBRID")

```

```python
m.add_raster(vis, layer_name='True Color')
m.add_raster(tir, colormap='terrain', layer_name='TIR')
```
